# **Quality Problems**

## **Completitud**

```python
def verificar_valores_nulos(data_frame):
    columnas_con_nulos = data_frame.columns[data_frame.isnull().any()].tolist()
    if not columnas_con_nulos:
        print("No hay valores nulos en el conjunto de datos.")
    else:
        print("Columnas con valores nulos:")
        for columna in columnas_con_nulos:
            print(f"- {columna}")

verificar_valores_nulos(df)
```

##**Consistencia**

```python
matriz_correlacion = df.corr()
print(matriz_correlacion)
plt.figure(figsize=(40, 40))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlación')
plt.show()
```
